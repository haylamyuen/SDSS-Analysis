### What We Did

<div style="

**Part 1 — Meeting the Data**

We introduced the SDSS and the physical meaning of every column in the dataset, from the five photometric bands to the Petrosian radius to the `zWarning` flag. We learned what redshift measures, why it is our primary distance proxy, and how it connects observation to cosmic time. We identified the three classes of objects in the dataset (stars, galaxies, and quasars), and understood why galaxies are our focus.

<br>

**Part 2 — Colour Bimodality**

We discovered the most fundamental observational fact in galaxy evolution: galaxies do not form a smooth continuum of colours. They cluster into two sharply distinct populations: the **Blue Cloud** of actively star-forming galaxies and a **Red Sequence** of quenched, passive galaxies, separated by a sparsely populated **Green Valley**.

<br>

**Part 3 — Structure and Morphology**

We asked whether a galaxy's *shape* tells the same story as its colour, and found that it largely does, but with important nuance. Blue galaxies tend to be disc-dominated spirals, while red galaxies tend to be bulge-dominated ellipticals. But the off-diagonal populations were the most interesting: **passive discs** (red but structurally disc-like) and **compact star-formers** (blue but concentrated) show that colour and morphology don't always match. Quenching does not always require structural transformation, and a galaxy can stop forming stars without destroying its disc.

<br>

**Part 4 — Distances, Sizes, and Luminosities**

We converted redshifts into physical distances using the $\Lambda\text{CDM}$ cosmological model, and used those distances to compute absolute magnitudes and physical sizes. This revealed that quenching is mass-dependent: the most luminous and massive galaxies are almost universally red and quenched. Also, spirals are physically larger than ellipticals of the same mass, and at any fixed luminosity, blue star-forming galaxies are roughly 1.5–2× larger than red quenched ones.

<br>

**Part 5 — Machine Learning**

We applied two machine learning techniques to the dataset:

* **Photometric redshift prediction (supervised regression):** A Random Forest trained on galaxy colour indices predicted spectroscopic redshifts with high accuracy. The $g−i$ colour index, spanning the widest useful wavelength range, carried by far the most information.

* **Lifecycle positioning (unsupervised PCA + K-Means):** Without being told anything about galaxy types or the Hubble sequence, the algorithm independently discovered five evolutionary groups: Active Spirals, Transition Spirals, Green Valley galaxies, Passive Discs, and Ellipticals, which match the physical picture almost exactly. The first and second principal components explained the largest fraction of variance in the data, and aligned with the colour-morphology evolutionary axis, while the third and fourth components represented a more structural and geometric picture of galaxies.

### Answering the Central Question

---

<div style="width: 1200px">

#### What drives galaxy transformation?

From our analysis, the data points consistently to **stellar mass** as the primary driver of galaxy transformation. At every point of the analysis (colour bimodality, size-luminosity relations, quenched fractions across cosmic time, machine learning feature importances), the same pattern was found: more massive galaxies are more evolved. They are redder, more compact, more bulge-dominated, and quite a bit more likely to have quenched their star formation.

However, the *mechanism* behind mass-dependent quenching is less certain, and it probably involves several processes working together:

* **AGN feedback**: Supermassive black holes at the centres of massive galaxies become active and blast out energy that heats and expels the cold gas needed for star formation. Since black hole mass scales with galaxy mass, this mechanism naturally targets the most massive systems.

* **Gas strangulation**: Here, the supply of fresh cold gas from the surrounding cosmic web is cut off, often after a major merger. Without new fuel, star formation slowly exhausts the remaining gas and dies.

* **Ram-pressure stripping**: In this process, galaxies moving through the hot gas of galaxy clusters have their own cold gas stripped away by the pressure of the surrounding medium, quenching them rapidly.

<br>

#### Can we predict where a galaxy is in its evolutionary lifecycle?

**Yes.** Definitely. By finding the direction in which galaxy feature change, our unsupervised model was able to compare where different galaxies sit in their evoltionary lifecycle. This axis of variation:

* Correlates tightly with $g-r$ colour
* Encodes structural information independently of colour
* Was discovered entirely without prior knowledge

So, galaxy evolution, despite involving billions of stars developing across millions of light-years, and black holes of unimaginable mass, ultimately reduces to a surprisingly **low-dimensional process**. Most of the variation in the galaxy population can be captured by just two numbers: where a galaxy sits on the evolutionary axis, and how its structure has responded to that evolution.

### The Bigger Picture

---

<div style="width: 1200px">

This project has been a demonstration of how modern astronomy works: not by pointing a telescope and writing down what you see, but by combining enormous datasets, careful physical reasoning, and computational tools to extract meaning from numbers.

Every result we found - the bimodality, the mass-quenching relationship, the size offset between spirals and ellipticals, the lifecycle positioning - was already present in the SDSS dataset from the moment it was collected. What we did was learn how to read it.

There's still heaps that remains unresolved. We don't know which quenching mechanism dominates, or whether different mechanisms operate in different mass regimes or environments. We don't know exactly when the Red Sequence formed, and whether it was already in place at $z = 2$, or whether it has been growing steadily since. We can't directly observe the quenching event itself, only the before and after.

These are the questions driving the next generation of surveys, like **Euclid**, **LSST**, and the **Square Kilometre Array**, which will observe billions of galaxies across the full history of the universe. And who knows? The methods developed in this project, like photometric redshift prediction, PCA-based lifecycle mapping, and unsupervised clustering, might just be the tools that astronomers use to discover something new.
